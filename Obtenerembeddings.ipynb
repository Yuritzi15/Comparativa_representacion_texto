```python

url = "https://raw.githubusercontent.com/Yuritzi15/Word2Vec/refs/heads/main/data/Requerimientos.csv"
df = pd.read_csv(url)
df['class'] = df['class'].replace({"fr": 0, "nfr": 1})
df.head()
```

```python
#Cambia los valores a False o True y compara los resultados
preprocesador=Preprocesamiento(
    aplicar_lowercase=True,
    remover_stopwords=True,
    aplicar_stemming=True,
    aplicar_lematizacion=False
)

#aplica el preprocesamiento
df['tokens']=df['text'].apply(preprocesador.procesamiento)

#crea el vocabulario
todos_los_tokens = []
for lista_tokens in df['tokens']:
  todos_los_tokens.extend(lista_tokens)
vocabulario=sorted(set(todos_los_tokens))

print("Vocabulario:",vocabulario)
print("Tamaño del vocabulario:",len(vocabulario))
```

#Obtener embeddings CBOW

```python
import requests
class CBOW:
    def __init__(self,cbow_embeddings_url, cbow_vocab_url):
        self.cbow_embeddings_url = cbow_embeddings_url
        self.cbow_vocab_url = cbow_vocab_url
        self.embeddings, self.vocab = self.cargar_embeddings()

    def cargar_embeddings(self):
        cbow_embeddings = np.loadtxt(self.cbow_embeddings_url, delimiter='\t')
        response = requests.get(self.cbow_vocab_url)
        response.raise_for_status()  # para lanzar error si algo falla
        vocab_lines = response.text.strip().splitlines()
        cbow_vocabulario = {word: idx for idx, word in enumerate(vocab_lines)}
        return cbow_embeddings, cbow_vocabulario

    def generardoc2vec(self, tokens):

      vecs = [self.embeddings[self.vocab[word]] for word in tokens if word in self.vocab]
      if len(vecs) == 0:
        return np.zeros(self.embeddings.shape[1])
      return np.sum(vecs, axis=0)
```

```python
#importar embeddings y vocabulario

cbow_embeddings_url = "https://raw.githubusercontent.com/Yuritzi15/Comparativa_representacion_texto/refs/heads/main/Word2Vec/cbow_embeddings/embeddings_cbow.tsv"


cbow=CBOW(cbow_embeddings_url, cbow_vocab_url)
cbow_embeddings, cbow_vocabulario = cbow.cargar_embeddings()

df['cbow_embedding'] = df['tokens'].apply(lambda t: cbow.generardoc2vec(t))
# Revisar resultado
df['cbow_embedding'][0].shape
```

#Obtener embedding FastText

```python
class FASTTEXT:
    def __init__(self,fasttext_embeddings_url, fasttext_vocab_url):
        self.fasttext_embeddings_url = fasttext_embeddings_url
        self.fasttext_vocab_url = fasttext_vocab_url
        self.embeddings, self.vocab = self.cargar_embeddings()

    def cargar_embeddings(self):
        fasttext_embeddings = np.loadtxt(self.fasttext_embeddings_url, delimiter='\t')
        response = requests.get(self.fasttext_vocab_url)
        response.raise_for_status()  # para lanzar error si algo falla
        vocab_lines = response.text.strip().splitlines()
        fasttext_vocabulario = {word: idx for idx, word in enumerate(vocab_lines)}
        return fasttext_embeddings, fasttext_vocabulario

    def generar_n_gramas(self, word, n_min=3, n_max=6):
      n_grams = []
      word = '<' + word + '>'
      for n in range(n_min, n_max + 1):
          for i in range(len(word) - n + 1):
              n_grams.append(word[i:i+n])
      return n_grams

    def embedding_palabra(self, word, n_min=3, n_max=6):
      ngrams = self.generar_n_gramas(word, n_min, n_max)
      vectors = [self.embeddings[self.vocab[ng]] for ng in ngrams if ng in self.vocab]
      if vectors:
          return np.sum(vectors, axis=0)
      else:
          return np.zeros(self.embeddings.shape[1])

    def generardoc2vec(self, tokens, n_min=3, n_max=6):
      vecs = []
      word_vectors = [self.embedding_palabra(word, n_min, n_max) for word in tokens]
      word_vectors = [vec for vec in word_vectors if vec is not None and np.linalg.norm(vec) > 0]
      if word_vectors:
        return np.sum(word_vectors, axis=0)
      else:
        return np.zeros(self.embeddings.shape[1])
```

```python
#importar embeddings y vocabulario

fasttext_embeddings_url = "https://raw.githubusercontent.com/Yuritzi15/Comparativa_representacion_texto/refs/heads/main/FastText/fasttext_embeddings.tsv"
fasttext_vocab_url = "https://raw.githubusercontent.com/Yuritzi15/Comparativa_representacion_texto/refs/heads/main/FastText/fasttext_vocab.tsv"

fasttext=FASTTEXT(fasttext_embeddings_url, fasttext_vocab_url)
fasttext_embeddings, fasttext_vocab = fasttext.cargar_embeddings()
print("Shape de embeddings:", fasttext_embeddings.shape)
print("Primeras 5 palabras del vocabulario:", list(fasttext_vocab.items())[:5])

df['fasttext_embedding'] = df['tokens'].apply(lambda t: fasttext.generardoc2vec(t))
# # Revisar resultado
df['fasttext_embedding'][0].shape
```

#Obtener embeddings GloVe

```python
import numpy as np
import requests

class GLOVE:
    def __init__(self, glove_embeddings_url):
        self.glove_embeddings_url = glove_embeddings_url
        self.embeddings, self.vocab = self.cargar_embeddings()

    def cargar_embeddings(self):
        # Soporta tanto URL como ruta local
        if self.glove_embeddings_url.startswith("http"):
            response = requests.get(self.glove_embeddings_url)
            response.raise_for_status()
            lines = response.text.strip().split('\n')
        else:
            with open(self.glove_embeddings_url, 'r', encoding='utf-8') as f:
                lines = f.read().strip().split('\n')

        vocab = {}
        embedding_list = []

        for idx, line in enumerate(lines):
            parts = line.strip().split('\t')
            word = parts[0]
            vector = list(map(float, parts[1:]))
            vocab[word] = idx
            embedding_list.append(vector)

        embeddings = np.array(embedding_list)
        return embeddings, vocab

    def generardoc2vec(self, tokens):
        vecs = [self.embeddings[self.vocab[word]] for word in tokens if word in self.vocab]
        if len(vecs) == 0:
          return np.zeros(self.embeddings.shape[1])
        return np.sum(vecs, axis=0)
```

```python
#importar embeddings y vocabulario

glove_embeddings_url = "https://raw.githubusercontent.com/Yuritzi15/Comparativa_representacion_texto/refs/heads/main/GloVe/glove_embeddings.tsv"

glove=GLOVE(glove_embeddings_url)
glove_embeddings = glove.cargar_embeddings()
df['GloVe_embedding'] = df['tokens'].apply(lambda t: glove.generardoc2vec(t))
df['GloVe_embedding'][0].shape
```
